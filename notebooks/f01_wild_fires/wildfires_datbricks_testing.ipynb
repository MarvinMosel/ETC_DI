```python
import pyodbc 
import pandas as pd
import matplotlib.pyplot as plt

# Some other example server values are
cnxn  = pyodbc.connect('Driver={SQL Server};'
                        'Server=ONAGER;'
                        'Database=ETC_ULS_Sandbox;'
                        'UID=user1;'
                        'PWD=uSer1Pass!;'
                        'Trusted_Connection=yes;')
cursor = cnxn.cursor()


### SQL query
print ("Reading table from ONAGER")




query=('''
SELECT 
     wild_fires_20002022.[ADM_ID],ISO2,[LEVEL3_COD],TAA,
      
      
 ISNULL(SUM(y2022_jan), 0 ) AS  [01-01-2022],
 ISNULL(SUM(y2022_feb), 0 ) AS  [01-02-2022],
 ISNULL(SUM(y2022_mar), 0 )AS  [01-03-2022],
 ISNULL(SUM(y2022_apr), 0 ) AS [01-04-2022],
 ISNULL(SUM(y2022_may), 0 ) AS [01-05-2022],
 ISNULL(SUM(y2022_jun), 0 ) AS  [01-06-2022],
 ISNULL(SUM(y2022_jul), 0 )  As [01-07-2022],
 ISNULL(SUM(y2022_aug), 0 )  As [01-08-2022],
 ISNULL(SUM(y2022_sep), 0 ) AS [01-09-2022],
 ISNULL(SUM(y2022_oct), 0 ) AS  [01-10-2022],
 ISNULL(SUM(y2022_nov), 0 ) AS  [01-11-2022],
 ISNULL(SUM(y2022_dec), 0 ) AS  [01-12-2022]---,
 ----ISNULL( SUM([AreaHa]), 0 ) as Mapped_AreaHa
      
  FROM [ETC_ULS_Sandbox].[wild_fires].[wild_fires_20002022]

  left JOIN [ETC_ULS_Sandbox].[wild_fires].[LUT_adm_eea39_2021LUT] ON [LUT_adm_eea39_2021LUT].[ADM_ID]=[wild_fires_20002022].[ADM_ID]

 ---where [ISO2] = 'ES'
  
  group by wild_fires_20002022.ADM_ID, ISO2,[LEVEL3_COD],TAA


''')
df = pd.read_sql(query, cnxn)
#print(df.head(26))



#pivot 




print ("end.............")
```
